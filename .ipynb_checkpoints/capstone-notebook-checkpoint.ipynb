Reading the given wikipedia page

```python
import requests
website_url=requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
```

Importing the web scrapping tool BeautifulSoup and parse the url

```python
from bs4 import BeautifulSoup
```

```python
soup=BeautifulSoup(website_url,'html.parser')
```

```python
postalCodeList = []
boroughList = []
neighborhoodList = []
```

```python
for row in soup.find('table').find_all('tr'):
    cells = row.find_all('td')
    if(len(cells) > 0):
        postalCodeList.append(cells[0].text.rstrip('\n'))
        boroughList.append(cells[1].text.rstrip('\n'))
        neighborhoodList.append(cells[2].text.rstrip('\n'))
```

Creating the dataFrame

```python
toronto_df = pd.DataFrame({"PostalCode": postalCodeList,
                           "Borough": boroughList,
                           "Neighborhood": neighborhoodList})

toronto_df.head()
```

Dropping the not assigned terms 

```python
toronto_df_drop = toronto_df[toronto_df.Borough != "Not assigned"].reset_index(drop=True)
toronto_df_drop.head()
```

Grouping the table based on the postal code

```python
toronto_df_grouped=toronto_df_drop.groupby(['PostalCode','Borough'],as_index=False).agg(lambda x:", ".join(x))
toronto_df_grouped.head()
```

replacing the / by , in neighborhood

```python
toronto_df_grouped['Neighborhood']=toronto_df_grouped['Neighborhood'].str.replace('/',',')
```

```python
toronto_df_grouped.head()
```

Testing the completeness

```python
column_names = ["PostalCode", "Borough", "Neighborhood"]
test_df = pd.DataFrame(columns=column_names)

test_list = ["M5G", "M2H", "M4B", "M1J", "M4G", "M4M", "M1R", "M9V", "M9L", "M5V", "M1B", "M5A"]

for postcode in test_list:
    test_df = test_df.append(toronto_df_grouped[toronto_df_grouped["PostalCode"]==postcode], ignore_index=True)
test_df
```

Uisng foursquare API for locating nearby places

```python
venues=[]

for lat,long,post,borough,neighborhood in zip(toronto_df_new['Latitude'],toronto_df_new['Longitude'],toronto_df_new['PostalCode'],toronto_df_new['Borough'],toronto_df_new['Neighborhood']):
    url="https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}".format(CLIENT_ID,CLIENT_SECRET,VERSION,lat,long,radius,LIMIT)
    
    results=requests.get(url).json()["response"]["groups"][0]["items"]
    
    for venue in results:
        venues.append((
            post,
            borough,
            neighborhood,
            lat,
            long,
            venue['venue']['name'],
            venue['venue']['location']['lat'],
            venue['venue']['location']['lng'],
            venue['venue']['categories'][0]['name']))
```

Creating a new Dataframe with the details of the venue from the API and borough and postalcode from the previous dataframe

```python
venues_df=pd.DataFrame(venues)
venues_df.columns=['PostalCode','Borough','Neighborhood','BoroughLatitude','BoroughLongitude','VenueName','VenueLatitude','VenueLongitude','VenueCategory']

print(venues_df.shape)
venues_df.head()
```

```python
venues_df.groupby(['PostalCode','Borough','Neighborhood']).count()
```

generating one hot dataframe using get_dummies method from numpy on Category of venues

```python
toronto_onehot=pd.get_dummies(venues_df['VenueCategory'],prefix="",prefix_sep="")
toronto_onehot['PostalCode']=venues_df["PostalCode"]
toronto_onehot['Borough']=venues_df["Borough"]
toronto_onehot['Neighborhoods']=venues_df['Neighborhood']

fixed_columns=list(toronto_onehot.columns[-3:])+list(toronto_onehot.columns[:-3])
toronto_onehot=toronto_onehot[fixed_columns]

toronto_onehot.head()
```

Grouping the one hot datframe on the basis of its addres and taking mean of the venue category

```python
toronto_grouped=toronto_onehot.groupby(['PostalCode','Borough','Neighborhoods']).mean().reset_index()

print(toronto_grouped.shape)
toronto_grouped.head()
```

Sort the neiborhoods for the based on category of venue and generate the most common venue in a neighborhood

```python
num_top_venues=10

inicators=['st','nd','rd']

areaColumns=['PostalCode','Borough','Neighborhoods']
freqColumns=[]

for ind in np.arange(num_top_venues):
    try:
        freqColumns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        freqColumns.append('{}th Most Common Venue'.format(ind+1))
columns = areaColumns+freqColumns

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['PostalCode'] = toronto_grouped['PostalCode']
neighborhoods_venues_sorted['Borough'] = toronto_grouped['Borough']
neighborhoods_venues_sorted['Neighborhoods'] = toronto_grouped['Neighborhoods']

for ind in np.arange(toronto_grouped.shape[0]):
    row_categories = toronto_grouped.iloc[ind, :].iloc[3:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    neighborhoods_venues_sorted.iloc[ind, 3:] = row_categories_sorted.index.values[0:num_top_venues]

# neighborhoods_venues_sorted.sort_values(freqColumns, inplace=True)
print(neighborhoods_venues_sorted.shape)
neighborhoods_venues_sorted
```

Create a cluster and then group the Neighborhoods into different cluster

```python
kclusters=5

toronto_grouped_clustering=toronto_grouped.drop(['PostalCode','Borough','Neighborhoods'],axis=1)
kmean=KMeans(n_clusters=kclusters,random_state=0).fit(toronto_grouped_clustering)

kmean.labels_[0:10]
```

Cerate a new dataframe with each Neighborhood grouped into it's corresponding cluster

```python
toronto_merged=toronto_df_new.copy()
toronto_merged['Cluster Labels']=kmean.labels_
toronto_merged=toronto_merged.join(neighborhoods_venues_sorted.drop(["Borough", "Neighborhoods"], 1).set_index("PostalCode"), on="PostalCode")
print(toronto_merged.shape)
toronto_merged.head()
```

sort the table based on the label

```python
toronto_merged.sort_values(["Cluster Labels"], inplace=True)
toronto_merged
```

Map the Neighborhoods and color with different colors depending on it's cluster

```python
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=13)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i+x+(i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, post, bor, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['PostalCode'], toronto_merged['Borough'], toronto_merged['Neighborhood'], toronto_merged['Cluster Labels']):
    label = folium.Popup('{} ({}): {} - Cluster {}'.format(bor, post, poi, cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
